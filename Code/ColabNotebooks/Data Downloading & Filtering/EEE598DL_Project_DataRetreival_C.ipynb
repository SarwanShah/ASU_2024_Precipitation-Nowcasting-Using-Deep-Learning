# Downloading All Samples In A Specified Date Range


```python
# Finding All Urls in Specified Date Range
years = ['2014']
start_date = datetime.strptime("07", "%m")
end_date = datetime.strptime("09", "%m")
for year in years:
    current_date = start_date
    while current_date <= end_date:
        smonth = current_date.strftime('%m')
        print(f'Downloading for {smonth}/{year}')
        source_url = f"https://ghe.blob.core.windows.net/noaa-ghe/rain_rate/{year}/{smonth}"
        destination_path = f"/content/ghe_data/{year}/{smonth}"
        !mkdir -p "{destination_path}"
        !azcopy copy "{source_url}" "{destination_path}" --recursive
        current_date += relativedelta(months=1)  # Increment by one month
```
