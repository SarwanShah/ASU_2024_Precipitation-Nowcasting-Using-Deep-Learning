```python
from google.colab import drive
import os
import shutil

# Mount Google Drive
drive.mount('/content/drive')
root_dir = "/content/drive/My Drive/Graduate - Academic Work/Semester 3/EEE598 Deep Learning/Project/Data"

```

```python
# Define the base directory and subdirectories
base_dir = root_dir + '/Processed'
source_dir = base_dir + '/10mm'
dest_dir = base_dir + '/25mm'



# Define date arrays with format "YYYY/MM/DD"
dates_2024 = ['2024/07/01', '2024/07/02', '2024/07/03', '2024/07/04', '2024/07/05', '2024/07/07', '2024/07/08', '2024/07/09', '2024/07/10', '2024/07/11', '2024/07/12', '2024/07/13', '2024/07/14', '2024/07/15', '2024/07/16', '2024/07/17', '2024/07/18', '2024/07/19', '2024/07/20', '2024/07/21', '2024/07/22', '2024/07/23', '2024/07/24', '2024/07/25', '2024/07/27', '2024/07/28', '2024/07/29', '2024/07/30', '2024/07/31', '2024/08/01', '2024/08/02', '2024/08/03', '2024/08/04', '2024/08/05', '2024/08/06', '2024/08/07', '2024/08/08', '2024/08/09', '2024/08/10', '2024/08/12', '2024/08/13', '2024/08/14', '2024/08/15', '2024/08/16', '2024/08/17', '2024/08/18', '2024/08/20', '2024/08/21', '2024/08/22', '2024/08/23', '2024/08/24', '2024/08/25', '2024/08/26', '2024/08/27', '2024/08/28', '2024/08/29', '2024/08/30', '2024/08/31', '2024/09/01', '2024/09/02', '2024/09/03', '2024/09/04', '2024/09/05', '2024/09/07', '2024/09/08', '2024/09/26', '2024/09/27']

# Convert dates to "YYYYMMDD" format for comparison
def format_dates(dates):
    return [date.replace('/', '') for date in dates]

# Organize formatted dates by year
date_dict = {
    '2024': format_dates(dates_2024),
}

# Process files
for year, dates in date_dict.items():
    year_dir = source_dir + '/' + f'{year}'
    new_dist_dir = dest_dir + '/' + f'{year}'
    os.makedirs(new_dist_dir, exist_ok=True)
    print(year_dir)
    print(new_dist_dir)
    if os.path.exists(year_dir):  # Check if year directory exists
        for file_name in os.listdir(year_dir):
            # Extract the date from the file name (assumes consistent naming)
            file_date = file_name.split('.')[4][1:9]
            if file_date in dates:
                # Move matching file to the destination directory
                source_file = os.path.join(year_dir, file_name)
                dest_file = dest_dir + '/' + f'{year}' + '/' + f'{file_name}'

                shutil.copy2(source_file, dest_file)
                print(f"Copied: {file_name}")

print("File transfer complete!")
```
